```python
import matplotlib.pyplot as plt
plt.subplots(1, figsize=(10,10))

# plt.title('Receiver Operating Characteristic - DecisionTree')
plt.plot(myvar['MobileNet_fpr'][0], myvar['MobileNet_tpr'][0],color='blue',label='MobileNet')
plt.plot([0, 1], ls="--")
# plt.title('Receiver Operating Characteristic')
plt.plot(resnet_50['Resnet-50_fpr'][0], resnet_50['Resnet-50_tpr'][0],color='green',label='ResNet-50')
plt.plot(VGG_16['VGG_16_fpr'][0], VGG_16['VGG_16_tpr'][0],color='red',label='VGG 16')

plt.title('Receiver Operating Characteristic ')
plt.plot(VGG_19['VGG_19_fpr'][0], VGG_19['VGG_19_tpr'][0],color='aqua', label="VGG 19")
plt.plot(Fine_tuned_MobileNet['Fine_tuned_MobileNet_fpr'][2], Fine_tuned_MobileNet['Fine_tuned_MobileNet_tpr'][2],color='deeppink', label="Fine Tune MobileNet")
# plt.plot(test_fpr_sigmoid, test_tpr_sigmoid,color='brown', label="SVM sigmoid")
# plt.plot(test_fpr_rbf, test_tpr_rbf,color='peru', label="SVM rbf")

plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
## Calculating mean Precision of Five model
print(np.mean(myvar['MobileNet_precision']))
print(np.mean(resnet_50['Resnet-50_precision']))
print(np.mean(VGG_16['VGG_16_precision']))
print(np.mean(VGG_19['VGG_19_precision']))
print(np.mean(Fine_tuned_MobileNet['Fine_tuned_MobileNet_precision']))
```
